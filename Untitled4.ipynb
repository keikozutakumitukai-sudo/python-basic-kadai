```python
#   台形の各辺の長さ（㎝）
upper_base  = 10   #　上辺
lower_base  = 20   #  下辺
height  = 5       #高さ

# 台形の面積の計算
area  = (upper_base + lower_base) * height / 2

#   計算結果の表示
print("台形の面積は", area,"㎠です")
```
